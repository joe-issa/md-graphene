```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= 2.
y=1.

l = 1.42
d = 1.20
h = 1.06

ly = np.sqrt(3)*l/2
lx = l/2
dy = np.sqrt(3)*d/2
dx = d/2
hy = np.sqrt(3)*h/2
hx = h/2

coords = [
            [0, 0, 0],
            [lx, ly, 0],
            [3*lx, ly, 0],
            [2*l, 0, 0],
            [3*lx, -ly, 0],
            [lx, -ly, 0],
            [3*l, 0, 0],
            [3*l+d, 0, 0],
            [4*l+d, 0, 0],
            [9*lx+d, ly, 0],
            [11*lx+d, ly, 0],
            [6*l+d, 0, 0],
            [11*lx+d, -ly, 0],
            [9*lx+d, -ly, 0],
            [7*l+d, 0, 0],
            [7*l+2*d, 0, 0],
            [8*l+2*d, 0, 0],
            [2*l, 2*ly, 0],
            [2*l+dx, 2*ly+dy, 0],
            [5*lx+dx, 3*ly+dy, 0],
            [7*lx+dx, 3*ly+dy, 0],
            [2*l, -2*ly, 0],
            [2*l+dx, -2*ly-dy, 0],
            [5*lx+dx, -3*ly-dy, 0],
            [7*lx+dx, -3*ly-dy, 0],
            [-2*l-d, 0, 0],

            [lx-hx, ly+hy, 0], 
            [-h, 0, 0],
            [lx-hx, -ly-hy, 0],
            [7*lx+dx+hx, 3*ly+dy-hy, 0],
            [7*lx+dx+hx, -3*ly-dy+hy, 0],
            [9*lx+d-hx, ly+hy, 0],
            [11*lx+d+hx, ly+hy, 0],
            [9*lx+d-hx, -ly-hy, 0],
            [11*lx+d+hx, -ly-hy, 0],
            [-2*l-d+h, 0, 0]
    
         ]

coords = np.array(coords)
```

```python
plt.plot(coords[:26,0], coords[:26,1], 'ko', label="Carbon")
plt.plot(coords[26:,0], coords[26:,1], 'bo', label="Hydrogen")
plt.xlabel("x (Å)")
plt.ylabel("y (Å)")
plt.legend(loc="best")
plt.title("A$\gamma$G-inspired material: the Unit-Cell")
plt.axis("square")
plt.savefig("uc.png")
```
